```python
# Step 3: Define segmentation logic
def segment_customer(row):
    volume = row["Transaction_Volume"]
    value = row["Transaction_Value"]
    
    # Determine transaction volume category
    if volume < 50:
        volume_category = "Low"
    elif 50 <= volume <= 200:
        volume_category = "Medium"
    else:
        volume_category = "High"

    # Determine transaction value category
    if value < 1000:
        value_category = "Low"
    elif 1000 <= value < 10000:
        value_category = "Medium"
    else:
        value_category = "High"

    segment = f"{row['Customer_Type']}_{row['Risk_Rating']}_{volume_category}_{value_category}"
    
    return volume_category, value_category, segment
merged_df = transaction_df.merge(account_df, on='Account_ID').merge(customer_df, on='Customer_ID')

# Step 1: Aggregate transaction data by Customer_ID
transaction_summary = merged_df.groupby("Customer_ID").agg(
    Transaction_Volume=("Transaction_ID", "count"),
    Transaction_Value=("Amount", "sum"),
    Avg_Transaction_Value=("Amount", "mean")
).reset_index()

# Step 2: Merge with customer data
segmented_df = customer_df.merge(transaction_summary, on="Customer_ID", how="left").fillna(0)
# Apply segmentation
segmented_df[["Volume_Category", "Value_Category", "Segment"]] = segmented_df.apply(segment_customer, axis=1, result_type="expand")
segmented_df.head(3)
# Display sample results
#print(segmented_df[["Customer_ID", "Customer_Type", "Risk_Rating", "Transaction_Volume", "Transaction_Value", "Segment"]].head(10))
```

```python
# Step 1: Join Tables
#merged_df = transaction_df.merge(account_df, on='Account_ID', how = 'inner').merge(segmented_df, on='Customer_ID', how = 'inner')

# Step 2: Identify Day 1 for each customer
min_dates = merged_df.groupby(['Customer_ID', 'tran_year', 'tran_month'])['Transaction_Date'].min().reset_index()
min_dates.rename(columns={'Transaction_Date': 'day1_date'}, inplace=True)

# Merge with transactions to mark Day 1 transactions
merged_df = merged_df.merge(min_dates, on=['Customer_ID', 'tran_year', 'tran_month'])
merged_df['day_diff'] = (merged_df['Transaction_Date'] - merged_df['day1_date']).dt.days

# Step 3: Compute WIRE Amount Debit on Day 1
wire_day1 = merged_df[
    (merged_df['Payment_Method'].isin(['Wire Transfer', 'Online Bank Transfer', 'ACH Transfer'])) & 
    (merged_df['DR_CR_Flag'] == 'Debit') & 
    (merged_df['day_diff'] == 0)
].groupby(['Customer_ID', 'tran_year', 'tran_month'])['Amount'].sum().reset_index()
wire_day1.rename(columns={'Amount': 'wire_debit_day1'}, inplace=True)

# Step 4: Filter Customers Where WIRE Debit on Day 1 > 0
valid_customers = wire_day1[wire_day1['wire_debit_day1'] > 0]['Customer_ID']
merged_df = merged_df[merged_df['Customer_ID'].isin(valid_customers)]

# Step 5: Compute 3-Day Cash Credit / Wire Debit Ratio
cash_credit_3d = merged_df[
    (merged_df['Payment_Method'].isin(['ATM', 'Bank Branch ATM', 'Bank Branch Cash'])) & 
    (merged_df['DR_CR_Flag'] == 'Credit') & 
    (merged_df['day_diff'] <= 5)
].groupby(['Customer_ID', 'tran_year', 'tran_month'])['Amount'].sum().reset_index()
cash_credit_3d.rename(columns={'Amount': 'cash_credit_3d'}, inplace=True)

wire_debit_3d = merged_df[
    (merged_df['Payment_Method'].isin(['Wire Transfer', 'Online Bank Transfer', 'ACH Transfer'])) & 
    (merged_df['DR_CR_Flag'] == 'Debit') & 
    (merged_df['day_diff'] <= 5)
].groupby(['Customer_ID', 'tran_year', 'tran_month'])['Amount'].sum().reset_index()
wire_debit_3d.rename(columns={'Amount': 'wire_debit_3d'}, inplace=True)

# Compute Ratio
ratio_df = cash_credit_3d.merge(wire_debit_3d, on=['Customer_ID', 'tran_year', 'tran_month'], how='outer').fillna(0)
ratio_df['cash_wire_ratio_3d'] = ratio_df['cash_credit_3d'] / ratio_df['wire_debit_3d']
ratio_df.replace([float('inf'), -float('inf')], 0, inplace=True)

# Step 6: Compute ABS(1 - CASH Credit / WIRE Debit) * 100
cash_credit_total = merged_df[
    (merged_df['Payment_Method'].isin(['ATM', 'Bank Branch ATM', 'Bank Branch Cash'])) & 
    (merged_df['DR_CR_Flag'] == 'Credit')
].groupby(['Customer_ID', 'tran_year', 'tran_month'])['Amount'].sum().reset_index()
cash_credit_total.rename(columns={'Amount': 'cash_credit_total'}, inplace=True)

wire_debit_total = merged_df[
    (merged_df['Payment_Method'].isin(['Wire Transfer', 'Online Bank Transfer', 'ACH Transfer'])) & 
    (merged_df['DR_CR_Flag'] == 'Debit')
].groupby(['Customer_ID', 'tran_year', 'tran_month'])['Amount'].sum().reset_index()
wire_debit_total.rename(columns={'Amount': 'wire_debit_total'}, inplace=True)

# Compute Deviation
deviation_df = cash_credit_total.merge(wire_debit_total, on=['Customer_ID', 'tran_year', 'tran_month'], how='outer').fillna(0)
deviation_df['cash_wire_deviation'] = abs(1 - deviation_df['cash_credit_total'] / deviation_df['wire_debit_total']) * 100
deviation_df.replace([float('inf'), -float('inf')], 0, inplace=True)

# Step 7: Combine All Metrics
final_df = wire_day1.merge(ratio_df, on=['Customer_ID', 'tran_year', 'tran_month'], how='inner').merge(deviation_df, on=['Customer_ID', 'tran_year', 'tran_month'], how='inner')

# Keep Only Customers Where WIRE Amount Debit on Day 1 > 0
final_df = final_df[final_df['wire_debit_day1'] > 0]


# ### ----------------- New ADD -----------------###
# # Step 8: Add Transaction IDs as a list per customer-month
# tx_ids = merged_df.groupby(['Customer_ID', 'tran_year', 'tran_month'])['Transaction_ID'].apply(list).reset_index()
# tx_ids.rename(columns={'Transaction_ID': 'transaction_ids'}, inplace=True)

# # Merge into final_df
# final_df = final_df.merge(tx_ids, on=['Customer_ID', 'tran_year', 'tran_month'], how='left')
# ### ----------------- ^ New ADD UP ^-----------------###

```

### 4-Change in Behavior (CIB) Activity - Average & Peak

```python
# Assuming transaction_df is already available and 'Transaction_Date' is datetime

# Step 1: Preprocessing
transaction_df['Transaction_Date'] = pd.to_datetime(transaction_df['Transaction_Date'])
transaction_df['tran_year'] = transaction_df['Transaction_Date'].dt.year
transaction_df['tran_month'] = transaction_df['Transaction_Date'].dt.month

# Step 2: Filter only Credit transactions
credit_txn = transaction_df[transaction_df['DR_CR_Flag'] == 'Credit']

# Step 3: Aggregate monthly credit amount
monthly_credit = credit_txn.groupby(['Customer_ID', 'tran_year', 'tran_month']).agg(
    Monthly_Credit_Amount=('Amount', 'sum')
).reset_index()

# Step 4: Create YearMonth for easier handling
monthly_credit['YearMonth'] = pd.to_datetime(monthly_credit['tran_year'].astype(str) + '-' + monthly_credit['tran_month'].astype(str) + '-01')

# Step 5: Sort by Customer_ID and YearMonth
monthly_credit = monthly_credit.sort_values(['Customer_ID', 'YearMonth'])

# Step 6: Rolling 12-month calculations
def rolling_calculations(df):
    df = df.set_index('YearMonth')  # Set datetime index
    df = df.sort_index()

    # Rolling 12M Sum
    df['Volume_Credit_12M'] = df['Monthly_Credit_Amount'].rolling(window='365D', min_periods=1).sum()

    # Rolling 12M Active Months
    df['Active_Months_12M'] = df['Monthly_Credit_Amount'].rolling(window='365D', min_periods=1).apply(lambda x: (x > 0).sum())

    # Avg = Total Amount / Active Months
    df['Avg_Credit_12M'] = df['Volume_Credit_12M'] / df['Active_Months_12M']
    
    # Maximum credit amount in last 12 months 
    df['Max_Credit_12M'] = df['Monthly_Credit_Amount'].rolling(window='365D', min_periods=1).max()

    # Rolling Std Dev
    df['Std_Credit_12M'] = df['Monthly_Credit_Amount'].rolling(window='365D', min_periods=1).std()

    # Percent Increase
    df['Percent_Increase_vs_Avg'] = ((df['Monthly_Credit_Amount'] - df['Avg_Credit_12M']) / df['Avg_Credit_12M']) * 100
    
    # Percent Increase
    df['Percent_Increase_vs_Peak'] = ((df['Monthly_Credit_Amount'] - df['Max_Credit_12M']) / df['Max_Credit_12M']) * 100

    # Z-Score
    df['Z_Score_Avg'] = (df['Monthly_Credit_Amount'] - df['Avg_Credit_12M']) / df['Std_Credit_12M']
    
    # Z-Score
    df['Z_Score_Peak'] = (df['Monthly_Credit_Amount'] - df['Max_Credit_12M']) / df['Std_Credit_12M']

    return df.reset_index()

# Step 7: Apply per Customer
final_detection_table = monthly_credit.groupby('Customer_ID').apply(rolling_calculations).reset_index(drop=True)

# Step 8: Final columns
final_detection_table = final_detection_table[['Customer_ID', 'YearMonth', 'Monthly_Credit_Amount',
                                                'Volume_Credit_12M', 'Active_Months_12M', 'Avg_Credit_12M', 
                                                'Max_Credit_12M', 'Std_Credit_12M', 'Percent_Increase_vs_Avg', 
                                                'Percent_Increase_vs_Peak', 'Z_Score_Avg', 'Z_Score_Peak']]

# View sample
final_detection_table.head()
```

## Features for ML model - 6 month lookback

```python
def rolling_unique_count(df, list_col_name, window=6):
    output_col = f'{list_col_name}_{window}_lookback_UniqueCount'
    df['month_date'] = pd.to_datetime(df['month_date'])
    df[output_col] = np.nan

    for cust_id, group in df.groupby('Customer_ID'):
        group = group.sort_values('month_date')
        rolling_values = []

        for i in range(len(group)):
            current_date = group.iloc[i]['month_date']
            window_start = current_date - pd.DateOffset(months=window)
            window_data = group[(group['month_date'] > window_start) & (group['month_date'] <= current_date)]

            # Flatten all Beneficiary_ID_List values and get unique count
            all_beneficiaries = set()
            for lst in window_data[list_col_name]:
                if isinstance(lst, list):
                    all_beneficiaries.update(lst)

            rolling_values.append(len(all_beneficiaries))

        df.loc[group.index, output_col] = rolling_values

    return df
def rolling_sum(df, column_name, window=6, count=False):
    op = 'Count' if count else 'Sum'
    output_col = f'{column_name}_{window}_lookback_{op}'
    df[output_col] = np.nan

    for cust_id, group in df.groupby('Customer_ID'):
        group = group.sort_values('month_date')
        values = []

        for i in range(len(group)):
            current_date = group.iloc[i]['month_date']
            window_start = current_date - pd.DateOffset(months=window)
            window_data = group[(group['month_date'] > window_start) & (group['month_date'] <= current_date)]

            if count:
                values.append(window_data[column_name].count())
            else:
                values.append(window_data[column_name].sum())

        df.loc[group.index, output_col] = values

    return df
def alert_enrichment(df_agg_merged):
    df = df_agg_merged.copy()
    df['tran_year'] = pd.to_numeric(df['tran_year'], errors='coerce')
    df['tran_month'] = pd.to_numeric(df['tran_month'], errors='coerce')
    df = df.dropna(subset=['tran_year', 'tran_month'])

    df['month_date'] = pd.to_datetime(df['tran_year'].astype(int).astype(str) + '-' + df['tran_month'].astype(int).astype(str) + '-01')

    # Unique Beneficiary_IDs per 6-month window
    df = rolling_unique_count(df, 'Beneficiary_ID_List', window=6)

    # Rolling aggregations
    df = df.sort_values(['Customer_ID', 'month_date'])

    df = rolling_sum(df, 'Monthly_Total_Tran_Amount', window=6, count=False)
    df = rolling_sum(df, 'Monthly_Debit_Tran_Amount', window=6, count=False)
    df = rolling_sum(df, 'Monthly_Credit_Tran_Amount', window=6, count=False)
    df = rolling_sum(df, 'Monthly_Total_Tran_Amount', window=6, count=True)
    df = rolling_sum(df, 'Monthly_Debit_Tran_Amount', window=6, count=True)
    df = rolling_sum(df, 'Monthly_Credit_Tran_Amount', window=6, count=True)
    df = rolling_sum(df, 'Monthly_HR_Tran_Amount', window=6, count=False)
    df = rolling_sum(df, 'Monthly_HR_Tran_Count', window=6, count=True)

    # Ratios and derived metrics
    df['%_6M_HRG_Tran_Amoun'] = (
        df['Monthly_HR_Tran_Amount_6_lookback_Sum'] / df['Monthly_Total_Tran_Amount_6_lookback_Sum']
    ).replace([np.inf, -np.inf], np.nan)

    df['%_6M_HRG_Tran_Count'] = (
        df['Monthly_HR_Tran_Count_6_lookback_Count'] / df['Monthly_Total_Tran_Amount_6_lookback_Count']
    ).replace([np.inf, -np.inf], np.nan)

    df['Credit_to_Debit_Amount_Ratio_6M'] = (
        df['Monthly_Credit_Tran_Amount_6_lookback_Sum'] / df['Monthly_Debit_Tran_Amount_6_lookback_Sum']
    ).replace([np.inf, -np.inf], np.nan)

    df['Credit_to_Debit_Count_Ratio_6M'] = (
        df['Monthly_Credit_Tran_Amount_6_lookback_Count'] / df['Monthly_Debit_Tran_Amount_6_lookback_Count']
    ).replace([np.inf, -np.inf], np.nan)

    return df
```

```python
import numpy as np
import pandas as pd
############################################################  last
def Alert_Flag_all_in_one_clipped_with_case_monthly(
    senarios_triggered_df, 
    risk_score_per_flag=15,  
    min_total_risk=45
):
    """
    Processes alerts to calculate total risk score per customer per month,
    generates unique Alert IDs, and assigns Case IDs using a 3-month rolling window logic.
    """
    def month_name_to_number(month_name):
        month_map = {
            'January': 1, 'February': 2, 'March': 3, 'April': 4, 
            'May': 5, 'June': 6, 'July': 7, 'August': 8, 
            'September': 9, 'October': 10, 'November': 11, 'December': 12
        }
        return month_map.get(month_name, month_name)

    # Step 1: Clean and prepare the data
    df = senarios_triggered_df.copy()
    df['tran_month'] = df['tran_month'].apply(month_name_to_number)
    df = df[df['tran_month'].notna()]
    df['tran_month'] = df['tran_month'].astype(int)
    df['tran_year'] = df['tran_year'].astype(int)  # <-- Fix added here
    df['event_IDs'] = df['event_IDs'].astype(str)  # -----------------------   added

    # Step 2: Get alert flags and calculate total risk score
    alert_cols = [col for col in df.columns if col.startswith('Alert_Flag_')]
    df['Total_Risk_Score'] = df[alert_cols].sum(axis=1) * risk_score_per_flag

    # Step 3: Aggregate by customer and month
    agg_df = df.groupby(['Customer_ID', 'tran_year', 'tran_month','event_IDs'], as_index=False)['Total_Risk_Score'].sum()
    agg_df = agg_df[agg_df['Total_Risk_Score'] >= min_total_risk]

    # Step 4: Generate unique Alert IDs
    agg_df['Alert_id'] = agg_df.apply(
        lambda row: f"{str(row['tran_month']).zfill(2)}_{row['tran_year']}_{row['Customer_ID']}_{np.random.randint(10, 100)}",
        axis=1
    )

    # Step 5: Assign Case_ID based on last 3 months logic
    agg_df.sort_values(by=['Customer_ID', 'tran_year', 'tran_month','event_IDs'], inplace=True)
    agg_df['Case_ID'] = None
    case_counter = 1
    customer_cases = {}

    for idx, row in agg_df.iterrows():
        cust = row['Customer_ID']
        year = row['tran_year']
        month = row['tran_month']
        current_period = year * 12 + month  # Convert to a comparable integer

        if cust not in customer_cases:
            # First case for this customer
            case_id = f"CASE_{case_counter:06d}"
            customer_cases[cust] = [(current_period, case_id)]
            agg_df.at[idx, 'Case_ID'] = case_id
            case_counter += 1
        else:
            # Check if any previous alert is within the last 3 months
            found_recent = False
            for prev_period, prev_case_id in reversed(customer_cases[cust]):
                if current_period - prev_period <= 3:
                    agg_df.at[idx, 'Case_ID'] = prev_case_id
                    customer_cases[cust].append((current_period, prev_case_id))
                    found_recent = True
                    break
            if not found_recent:
                case_id = f"CASE_{case_counter:06d}"
                customer_cases[cust].append((current_period, case_id))
                agg_df.at[idx, 'Case_ID'] = case_id
                case_counter += 1

    return agg_df

case_table = Alert_Flag_all_in_one_clipped_with_case_monthly(
    senarios_triggered_df, 
    risk_score_per_flag=15, 
    min_total_risk=45
)
```

### Case Priority Logic

```python
# --- Step 1: Define scoring logic ---
def score_risk(row):
    score = 0

    # Risk Rating
    if row['Risk_Rating'] == 'High':
        score += 3
    elif row['Risk_Rating'] == 'Medium':
        score += 2
    elif row['Risk_Rating'] == 'Low':
        score += 1

    # PEP flag
    if row['PEP_Flag'] == 'Yes' or row['Entity_PEP_Flag'] == 'Yes':
        score += 2
    # Total Alert Total_Risk_Score
    
    if row['Total_Risk_Score'] == 5*15:
        score += 3
    elif row['Total_Risk_Score'] == 4*15:
        score += 2
    elif row['Total_Risk_Score'] == 3*15:
        score += 1

    # Sanction body weight
    '''
    if row['Entity_Sanctioning_body'] == 'OFAC':
        score += 3
    elif row['Entity_Sanctioning_body'] == 'FATF':
        score += 2
    elif row['Entity_Sanctioning_body'] == 'EU':
        score += 1
    elif row['Entity_Sanctioning_body'] == 'UN':
        score += 2
    

    # List type weight
    if row['Entity_List_Type'] in ['Terrorist', 'Arms Dealing']:
        score += 3
    elif row['Entity_List_Type'] in ['Money Laundering', 'Fraud']:
        score += 2
    elif row['Entity_List_Type'] == 'PEP':
        score += 1
    '''
    return score

# --- Step 2: Apply scoring to each row ---
df['row_score'] = df.apply(score_risk, axis=1)

# --- Step 3: Aggregate to Case_ID level ---
case_scores = df.groupby(['Customer_ID','Case_ID']).agg(
    total_score=('row_score', 'sum'),
    alerts=('Alert_id', 'nunique'),#Counts unique Alert_ids per case.
    case_count = ('Case_ID','count'), #Counts total rows in each (Customer_ID, Case_ID) group.
  
    risk_rating=('Risk_Rating', 'first')  # Optional
).reset_index()

# --- Step 4: Assign priority band based on score ---
def assign_priority(score):
    if score >= 8:
        return 'High'
    elif score >= 5:
        return 'Medium'
    else:
        return 'Low'

case_scores['Priority'] = case_scores['total_score'].apply(assign_priority)

# --- Output ---
#print(case_scores[['Case_ID', 'total_score', 'alerts', 'Priority']])
case_scores
```

### Adding L_3 Flag

```python
# Ensure reproducibility
np.random.seed(42)

# Create a copy to avoid changing the original
case_scores_with_l3 = case_scores.copy()
case_scores_with_l3['L_3_flag'] = '-'  # default 

# Group by year and month
grouped = case_scores_with_l3.groupby(['tran_year', 'tran_month'])

for (year, month), group in grouped:
    n_cases = len(group)
    
    # Randomly choose 10-15% of the cases
    sample_frac = np.random.uniform(0.10, 0.15)
    n_l3 = int(np.ceil(n_cases * sample_frac))
    
    # Apply weighted sampling: 70% High, 20% Medium, 10% Low
    high_cases = group[group['Priority'] == 'High']
    medium_cases = group[group['Priority'] == 'Medium']
    low_cases = group[group['Priority'] == 'Low']
    
    n_high = int(n_l3 * 0.7)
    n_medium = int(n_l3 * 0.2)
    n_low = n_l3 - n_high - n_medium
    
    l3_ids = []

    l3_ids += list(high_cases.sample(min(n_high, len(high_cases)), random_state=42)['Case_ID'])
    l3_ids += list(medium_cases.sample(min(n_medium, len(medium_cases)), random_state=42)['Case_ID'])
    l3_ids += list(low_cases.sample(min(n_low, len(low_cases)), random_state=42)['Case_ID'])

    # Mark selected Case_IDs as 'L3'
    case_scores_with_l3.loc[
        case_scores_with_l3['Case_ID'].isin(l3_ids) &
        (case_scores_with_l3['tran_year'] == year) &
        (case_scores_with_l3['tran_month'] == month),
        'L_3_flag'
    ] = 'L3'    # I keep it as string to add 'L2' or 'L1' label in future if needed
case_scores_with_l3
```

```python
# --- Step 1: Define scoring logic ---
def score_risk(row):
    score = 0

    # Risk Rating
    if row['Risk_Rating'] == 'High':
        score += 3
    elif row['Risk_Rating'] == 'Medium':
        score += 2
    elif row['Risk_Rating'] == 'Low':
        score += 1

    # PEP flag
    if row['PEP_Flag'] == 'Yes' or row['Entity_PEP_Flag'] == 'Yes':
        score += 2
    # Total Alert Total_alert count
    
    if row['Alerts_count'] >= 3:
        score += 3
    elif row['Alerts_count'] == 2:
        score += 2
    elif row['Alerts_count'] == 1:
        score += 1

    # Sanction body weight
    '''
    if row['Entity_Sanctioning_body'] == 'OFAC':
        score += 3
    elif row['Entity_Sanctioning_body'] == 'FATF':
        score += 2
    elif row['Entity_Sanctioning_body'] == 'EU':
        score += 1
    elif row['Entity_Sanctioning_body'] == 'UN':
        score += 2
    

    # List type weight
    if row['Entity_List_Type'] in ['Terrorist', 'Arms Dealing']:
        score += 3
    elif row['Entity_List_Type'] in ['Money Laundering', 'Fraud']:
        score += 2
    elif row['Entity_List_Type'] == 'PEP':
        score += 1
    '''
    return score

# --- Step 2: Apply scoring to each row ---
df_ML['row_score'] = df_ML.apply(score_risk, axis=1)

# --- Step 3: Aggregate to Case_ID level ---
ML_case_scores = df_ML.groupby(['Customer_ID','Case_ID']).agg(
    total_score=('row_score', 'sum')
#     ,
#     alerts=('Alert_id', 'nunique'),#Counts unique Alert_ids per case.
#     case_count = ('Case_ID','count'), #Counts total rows in each (Customer_ID, Case_ID) group.
  
#     risk_rating=('Risk_Rating', 'first')  # Optional
).reset_index()

# --- Step 4: Assign priority band based on score ---
def assign_priority(score):
    if score >= 8:
        return 'High'
    elif score >= 5:
        return 'Medium'
    else:
        return 'Low'

ML_case_scores['Priority'] = ML_case_scores['total_score'].apply(assign_priority)

# --- Output ---
#print(case_scores[['Case_ID', 'total_score', 'alerts', 'Priority']])
ML_case_scores
```

```python
import pandas as pd
import pyarrow.parquet as pq
import json
import os

save_dir = r'C:/Users/sajad/Documents/Prepaid/Transaction Monitoring/New_tables/final_user_vars_V6.1'

# Load metadata
with open(os.path.join(save_dir, 'metadata.json'), 'r') as f:
    metadata = json.load(f)

loaded_dfs = {}

for var_name, meta in metadata.items():
    if meta['type'] == 'dataframe':
        file_path = meta['file']
        print(f"Loading {var_name} from {file_path}...")

        # Try reading in row-group chunks
        try:
            parquet_file = pq.ParquetFile(file_path)
            chunk_list = []
            for i in range(parquet_file.num_row_groups):
                chunk = parquet_file.read_row_group(i).to_pandas()
                chunk_list.append(chunk)
            loaded_dfs[var_name] = pd.concat(chunk_list, ignore_index=True)

        except Exception as e:
            print(f"Error reading {var_name}: {e}")
            loaded_dfs[var_name] = None  # Handle failure gracefully

# Restore to global namespace if needed
globals().update(loaded_dfs)
```
